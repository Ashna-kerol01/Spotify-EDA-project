```python
explicit_counts = df["explicit"].value_counts()

plt.figure(figsize=(5,5))
plt.pie(explicit_counts, labels=["Non-Explicit", "Explicit"], autopct="%1.1f%%")
plt.title("Explicit vs Non-Explicit Tracks")
plt.show()

# Popularity comparison
sbn.boxplot(x="explicit", y="track_popularity", data=df)
plt.title("Track Popularity by Explicit Content")
plt.show()
```

### üì¶ Popularity Distribution & Variability
