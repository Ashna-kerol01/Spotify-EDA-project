## Data Type Conversion: Album Release Date

The **`album_release_date`** column was initially stored as an object. The release year was extracted from it and converted into an **integer (`int`)** to make the data suitable for analysis.


```python
df["album_release_date"] = pd.to_datetime(df["album_release_date"] ,errors = "coerce")
```
